```python
def next_n(n):
    return n/2 if n%2==0 else 3*n+1
```

```python
import numpy as np
```

```python
n=27
s=[]
while n != 1:
    n=next_n(n)
    s.append(n)
def calc_s(n):
    s=[]
    while n != 1:
        n=next_n(n)
        s.append(n)
    return s, len(s), np.max(s)
```

```python
print(calc_s(49))
```

```python
import seaborn as sns
import pandas as pd
sns.set()
```

```python
ps = pd.DataFrame(s, columns=['n'])
sns.relplot(data=ps, kind='line')
```

```python
sns.distplot(ps, bins=50, kde=False)
```

```python
lower = 2
upper = 10000

primes=[]
for num in range(lower,upper + 1):
    if num > 1:
       for i in range(2,num):
           if (num % i) == 0:
               break
       else:
           primes.append(num)
```

```python
all = primes #range(2,10000)
r = [(n, calc_s(n)) for n in all]
r = [[e[0], e[1][1], e[1][2]] for e in r]
```

```python
res = pd.DataFrame(r, columns=['x', 'N', 'M'])
res = res[['x', 'N']]
```
